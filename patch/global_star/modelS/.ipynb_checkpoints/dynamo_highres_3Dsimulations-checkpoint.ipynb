## Plot Perturbations

#### plotting routine

```python
def plot_2D_perturbations(output):
    """ Simple routine to make 2D plots of different quantities """
    # Constants
    rsun = 695.7 # Mm
    boxsize = output.dx*(2**output.levelmax)
    # Initialize figure
    fig, axes = plt.subplots(3,2, figsize=(8,12))
    # Central slice index
    central_idx = int((2**output.levelmax)/2)
    # Compute perturbations
    d_prime = (np.abs(output.d - output.deq)/output.deq)[central_idx]
    p_prime = (np.abs(output.p - output.peq)/output.peq)[central_idx]
    s_prime = (np.abs(output.s - output.seq)/output.seq)[central_idx]
    e_prime = (np.abs(output.etot - output.etoteq)/output.etoteq)[central_idx]
    v_prime = (np.abs(output.v.norm/np.sqrt(1.6667*output.peq/output.deq)))[central_idx]
    B_prime = np.abs(output.B.norm)[central_idx]
    primes = [d_prime, p_prime, s_prime, e_prime, v_prime, B_prime]
    labels = ['density', 'pressure', 'entropy', 'energy', 'Mach number', 'B field']
    for ax,q,title in zip(axes.ravel(), primes, labels):
        # Compute vmax and vim
        #vmax = np.max(np.log10(q))
        #vmin = np.min(np.log10(q))

        # Make imshow
        cs = ax.imshow(np.log10(q).T, 
                       origin='lower',
                       #vmax=vmax,
                       #vmin=vmin,
                       extent=[-boxsize/2,boxsize/2,-boxsize/2,boxsize/2],
                       cmap='nipy_spectral')
        # Plot circles of different regions
        boundaries = [4.86990e2, 6.60915e2]
        for b in boundaries:
            interface = plt.Circle((0, 0), b, color='k', fill=None)
            ax.add_patch(interface)

        # Colorbar
        divider = make_axes_locatable(ax)
        cax = divider.new_vertical(size = '3%', pad = 0.03)
        fig.add_axes(cax)
        fig.colorbar(cs, cax = cax, orientation = 'horizontal', label=title)
        cax.xaxis.set_ticks_position('top')
        cax.xaxis.set_label_position('top')
    
    return axes, fig
```

### Plot average profiles

#### Plotting routine

```python
def plot_2D_radial_profiles(output):
    """ 
    Averages radially the quantities of the output and compares them to
    the equilibrium profiles
    """
    # Constants
    rsun = 695.7 # Mm
    boxsize = output.dx*(2**output.levelmax)
    # Initialize figure
    fig, axes = plt.subplots(3,2, figsize=(8,12))
    # Get number of cells per dimension
    N = 2**output.levelmax
    dx = output.dx
    # Boundaries
    boundaries = [4.86990e2, 6.60915e2]
    # Build radial coordinate
    x,y,z = np.meshgrid(np.arange(N), np.arange(N), np.arange(N))
    x = x-(N-1)/2
    y = y-(N-1)/2
    z = z-(N-1)/2
    r = np.sqrt(x**2 + y**2 + z**2)
    r = np.array(r, dtype=int)
    r_Mm = np.arange(np.max(r)+1)/(np.max(r)+1)*boxsize/2*np.sqrt(3)
    # Get quantities to plot
    quantities = [output.d, output.p, output.s, output.etot, output.v.norm, output.B.norm]
    equilibriums = [output.deq, output.peq, output.seq, output.etoteq, np.zeros(output.v.norm.shape), np.zeros(output.B.norm.shape)]
    labels = ['density', 'pressure', 'entropy', 'energy', 'velocity', 'B field'] 
    mean_quantities = []
    for ax, q, eq, title in zip(axes.ravel(), quantities, equilibriums, labels):
        q_mean = []
        qeq_mean = []
        for i in np.arange(np.max(r)+1):
            q_mean.append(np.mean(q[r==i]))
            qeq_mean.append(np.mean(eq[r==i]))
        q_mean = np.array(q_mean)
        qeq_mean = np.array(qeq_mean)
        ax.plot(r_Mm, q_mean, color='blue')
        ax.plot(r_Mm, qeq_mean, color='black')
        for b in boundaries:
            ax.axvline(b, linestyle=':', color='gray')
        ax.set_title(title)
        ax.set_yscale('log')
        if title=='B field':
            ax.set_ylim([1e-5, 1e3])
        mean_quantities.append(q_mean)
    return mean_quantities, fig
```

```python
mean_profiles, fig = plot_2D_radial_profiles(output)
fig.tight_layout()
#fig.savefig('modelS_profiles.png', dpi=200)
```

## Plot perturbation around average value

```python
def compute_radial_average_grid2(q_mean_profile, r):
    r_range = np.arange(np.max(r))
    print(r.shape, r_range.shape, q_mean_profile.shape) 
    q_mean_grid = np.interp(r, r_range, q_mean_profile)
    return q_mean_grid


def compute_radial_average_grid(q, r):
    """ Computes a grid with the average radial value of a quantity at the given radius in each pixel """
    q_mean = np.zeros(r.shape)
    for r_target in np.arange(np.max(r)+1):
        # Compute the average
        mask = np.where(r==r_target)
        q_mean_radius = np.mean(q[mask])
        q_mean[mask] = q_mean_radius
    mask = np.where(r==0)
    q_mean[mask] = q[mask]
    return q_mean

def compute_radial_coordinate(levelmax):
    # Get number of cells per dimension
    N = 2**levelmax
    # Build radial coordinate
    x,y,z = np.meshgrid(np.arange(N), np.arange(N), np.arange(N))
    x = x-(N-1)/2
    y = y-(N-1)/2
    z = z-(N-1)/2
    r = np.sqrt(x**2 + y**2 + z**2)
    r = np.array(r)#, dtype=int)
    return r

def plot_2D_perturbations_around_average(output, mean_profiles):
    """ Simple routine to make 2D plots of different quantities """
    # Constants
    rsun = 695.7 # Mm
    boxsize = output.dx*(2**output.levelmax)
    # Central slice index
    central_idx = int((2**output.levelmax)/2)
    # Compute radial coordinates
    r = compute_radial_coordinate(output.levelmax)
    # Initialize figure
    fig, axes = plt.subplots(3,2, figsize=(8,12))
    # Compute perturbations
    quantities = [output.d, output.p, output.s, output.etot, output.v.norm, output.B.norm]
    labels = ['density', 'pressure', 'entropy', 'energy', 'velocity', 'B field']
    for ax,q,profile,title in zip(axes.ravel(), quantities, mean_profiles, labels):
        # Compute average grid
        q_mean = compute_radial_average_grid2(profile, r)
        q_prime = ((q - q_mean))[:,central_idx]
        # Compute vmax and vim
        # vmax = np.max(q_prime)
        # vmin = np.min(q_prime)

        # Make imshow
        cs = ax.imshow(q_prime.T, 
                       origin='lower',
                       # vmax=vmax,
                       # vmin=vmin,
                       extent=[-boxsize/2,boxsize/2,-boxsize/2,boxsize/2],
                       cmap='nipy_spectral')
        # Plot circles of different regions
        boundaries = [4.86990e2, 6.60915e2]
        for b in boundaries:
            interface = plt.Circle((0, 0), b, color='k', fill=None)
            ax.add_patch(interface)

        # Colorbar
        divider = make_axes_locatable(ax)
        cax = divider.new_vertical(size = '3%', pad = 0.03)
        fig.add_axes(cax)
        fig.colorbar(cs, cax = cax, orientation = 'horizontal', label=title)
        cax.xaxis.set_ticks_position('top')
        cax.xaxis.set_label_position('top')
    
    return axes, fig
```

```python
plot_2D_perturbations_around_average(output, mean_profiles)
```

## Energy analysis

```python
def internal_energy(output):
    """ Compute the internal energy density of an output """
    gamma = 5./3.
    ei = output.p/(gamma - 1.0)
    return ei

def kinetic_energy(output):
    """ Compute the kinetic energy density of an output """
    ekin = 0.5*output.d*(output.v.x**2 + output.v.y**2)
    return ekin

def magnetic_energy(output):
    """ Compute the magnetic energy density of an output """
    emag = 0.5*(output.B.x**2 + output.B.y**2)
    return emag

def total_energy(output):
    """ Computes the total energy density of an output and its components """
    ei = internal_energy(output)
    ekin = kinetic_energy(output)
    emag = magnetic_energy(output)
    etot = ei+ekin+emag
    energy = {'etot':etot, 'ei':ei, 'ekin':ekin, 'emag':emag}
    return energy

def compute_energy_evolution(n_outputs):
    """ Given a set of output numbers, it computes the average energy components of the outputs """
    energy_evolution = []
    # Loop
    for n in n_outputs:
        output = RamsesOutput()
        output.read(folder_data, n, norms=norms)
        energy_output = total_energy(output) 
        mean_values = np.array([np.mean(energy_output[k]) for k in energy_output.keys()])
        energy_evolution.append(mean_values)
    energy_evolution = np.array(energy_evolution)
    return energy_evolution


def compute_conv_zone_energy_evolution(n_outputs, levelmax):
    """ Given a set of output numbers, it computes the average energy components of the outputs """
    energy_evolution = []
    r = compute_radial_coordinate(levelmax)
    dx = 1.5E+03/2.**levelmax
    r_conv_up = int(6.60915e2/dx)
    r_conv_down = int(4.86990e2/dx)
    mask = np.where((r>=r_conv_down) & (r<=r_conv_up))
    # Loop
    for n in n_outputs:
        output = RamsesOutput()
        output.read(folder_data, n, norms=norms)
        energy_output = total_energy(output) 
        mean_values = np.array([np.mean(energy_output[k][mask]) for k in energy_output.keys()])
        energy_evolution.append(mean_values)
    energy_evolution = np.array(energy_evolution)
    return energy_evolution

def plot_energy_evolution(energy_evolution, only_etot=False):
    """ Simple plotting routine to plot the evolution in time of the energy components """
    # Get energy components
    etot = energy_evolution[:,0]
    ei   = energy_evolution[:,1]
    ekin = energy_evolution[:,2]
    emag = energy_evolution[:,3]
    # Prepare plot
    fig, ax = plt.subplots(figsize=(5,5))
    # Plot lines
    if not only_etot:
        ax.plot(ei  , label=r'$e_{i}$')
        ax.plot(ekin, label=r'$e_{kin}$')
        ax.plot(emag, label=r'$e_{mag}$')
    ax.plot(etot, label=r'$e_{tot}$', linestyle=':')
    ax.legend()
    ax.set_yscale('log')
    #ax.set_ylim([max(np.log10(etot))-1, max(np.log10(etot))+1])
    return fig, ax
```

#### Compute energies for a sequence of outputs and plot the evolution

```python
n_outputs = np.arange(1, 101, 10)
energy_evolution = compute_conv_zone_energy_evolution(n_outputs, 7)
```

```python
plot_energy_evolution(energy_evolution, False)
```
